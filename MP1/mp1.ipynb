This project will communicate the trends and percentiles of responses to a survey examining movie series

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("movieDF.csv")

df = pd.DataFrame(data)
```

```python
#function for getting the total number of observations in each given each column
def getPercentages(col):
    percentDict = {}
    colGroup = df.groupby(col).size()
    total = df.shape[0]
    for i in range(colGroup.size):
        percent = colGroup.iloc[i]*100/total
        percentDict[colGroup.index[i]] = round(percent,2)

    sortedDict = dict(sorted(percentDict.items(), key=lambda item: item[1], reverse=True))

    return sortedDict




## retrieve all percentages from each column
#who?
print(getPercentages("What is your age?"))
#who?
print(getPercentages("What is your major?"))
#who?
print(getPercentages("What state are you from?"))

#what?
print(getPercentages("What is the best movie series?"))

#when?
print(getPercentages("From the above, what age were you when you first watched a movie from this series ?"))


#where?
print(getPercentages("What streaming service did you watch this on?"))


#why?
print(getPercentages("Why is this the best series?"))


#why?
print(getPercentages("Why did you watch it at this age?"))


#why?
print(getPercentages("Should this movie first be seen in theaters? Why?"))
```

```python
#store all categories in age
ageGroup = df.groupby("What is your age?").size()

#make piechart
plt.rcParams["text.color"] = "black"
plt.rcParams["font.size"] = 18
plt.pie(x=ageGroup, 
           labels=ageGroup.index, 
           autopct = '%1.0f%%')
plt.show()
```
